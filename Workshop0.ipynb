Task 1 - Sum of Nested Lists:

```python
def sum_nested_list(nested_list):
    total = 0
    for item in nested_list:
        if isinstance(item, list):
            # Recursively sum the nested list
            total += sum_nested_list(item)
        else:
            # Add the number directly
            total += item
    return total

# Test with the sample nested list
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print("Total sum:", result)

```

Task 2 - Generate All Permutations of a String:

```python
def generate_permutations(s):
    # Base case: a single character string has only one permutation
    if len(s) <= 1:
        return [s]

    permutations = []
    for i, char in enumerate(s):
        # Take the current character out and find permutations of the remaining string
        remaining = s[:i] + s[i+1:]
        for p in generate_permutations(remaining):
            permutations.append(char + p)

    # Return unique permutations
    return list(set(permutations))

# Test with "abc"
print(generate_permutations("abc"))
# Test with "aab"
print(generate_permutations("aab"))
```

Task 3 - Directory Size Calculation:

```python
def calculate_directory_size(directory):
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            # If it's a subdirectory, recurse
            total_size += calculate_directory_size(value)
        else:
            # Otherwise, it's a file, add its size
            total_size += value
    return total_size
# Example directory structure
sample_directory = {
    "file1.txt": 120,
    "file2.txt": 80,
    "subdir1": {
        "file3.txt": 200,
        "subsubdir1": {
            "file4.txt": 50
        }  },
    "subdir2": {
        "file5.txt": 300
    }
}

# Test the function
total = calculate_directory_size(sample_directory)
print("Total directory size:", total, "KB")
```

8.2.1 Exercises - Dynamic Programming:
Task 1 - Coin Change Problem:

```python
def min_coins(coins, amount):
    # Initialize DP array with amount+1 (represents infinity)
    dp = [amount + 1] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins needed for amount 0

    # Fill the DP table
    for i in range(1, amount + 1):
        for coin in coins:
            if coin <= i:
                dp[i] = min(dp[i], dp[i - coin] + 1)

    # If dp[amount] is still "infinity", return -1
    return dp[amount] if dp[amount] != amount + 1 else -1

# Test
coins = [1, 2, 5]
amount = 11
print("Minimum coins needed:", min_coins(coins, amount))
```

Task 2 - Longest Common Subsequence (LCS):

```python
def longest_common_subsequence(s1, s2):
    m, n = len(s1), len(s2)
    # Create a DP table of size (m+1) x (n+1)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    # The length of LCS is in dp[m][n]
    return dp[m][n]

# Test
s1 = "abcdefghijkl"
s2 = "acegik"
print("Length of LCS:", longest_common_subsequence(s1, s2))
```

Task 3 - 0/1 Knapsack Problem:

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    # Create DP table of size (n+1) x (capacity+1)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Fill DP table
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i - 1] <= w:
                # Max of including or excluding the current item
                dp[i][w] = max(values[i - 1] + dp[i - 1][w - weights[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

# Test
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print("Maximum value achievable:", knapsack(weights, values, capacity))
```
